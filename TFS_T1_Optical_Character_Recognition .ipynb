# Optical Character Recognition (ORC)

```python
# import all library
import easyocr   # library for extract text, pre- instlled pytorch library required
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## A) Draw Result

```python
# import a input image
image_path = 'image.jpg' # load image
reader = easyocr.Reader(['en'], gpu = False) # extract ony english text

result = reader.readtext(image_path) # output result
```

```python
# display input original image
img = cv2.imread(image_path)
plt.imshow(img)
plt.show()
```

## B) Display Extracted Text

```python
# draw o rectangle box arround text

for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]]) 
    bottom_right = tuple([int(val) for val in detection[0][2]])
    cv2.rectangle(img, top_left, bottom_right, (0,255,0),1)
# display image with line in box
plt.figure(figsize =(5,5))
plt.imshow(img)
plt.show()

# print all line in image
for line in result:
    text = line[1]
    print(text)
```
